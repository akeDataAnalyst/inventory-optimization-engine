#### 1. Load Processed Data and Setup

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the "Intelligent" data
df = pd.read_csv('../data/processed/inventory_intelligence.csv')

# Quick check on the 'REORDER NOW' status
status_counts = df['Stock_Status'].value_counts()
print("Data Loaded. Current Inventory Health Summary:")
print(status_counts)
```

#### 2. The ABC Investment Treemap

```python
fig_abc = px.treemap(df, 
                     path=['ABC_Class', 'Category', 'Product_Name'], 
                     values='Annual_Consumption_USD',
                     color='ABC_Class',
                     color_discrete_map={
                         'A (High Value)': '#003366',
                         'B (Medium Value)': '#336699', 
                         'C (Low Value)': '#99CCFF'
                     },
                     title="Inventory Investment Portfolio (ABC Classification)")

fig_abc.update_traces(textinfo="label+value+percent parent")
fig_abc.show()
```

#### 3. Stockout Risk vs. Lead Time (The "Danger Zone")

```python
fig_risk = px.scatter(df, 
                      x="Lead_Time_Days", 
                      y="Initial_Stock",
                      size="Annual_Consumption_USD",
                      color="Stock_Status",
                      hover_name="Product_Name",
                      text="SKU",
                      title="Lead Time vs. Stock Level: Identifying Import Risks",
                      color_discrete_map={'REORDER NOW': 'red', 'WATCHLIST': 'orange', 'HEALTHY': 'green'},
                      labels={"Lead_Time_Days": "Import Lead Time (Days)", "Initial_Stock": "Current Units in Store"})

avg_rop = df['Reorder_Point'].mean()
fig_risk.add_hline(y=avg_rop, line_dash="dash", line_color="red", annotation_text="Average Reorder Trigger")

fig_risk.show()
```
